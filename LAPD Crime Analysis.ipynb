```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set a clean visual style
sns.set_theme(style="whitegrid")

def main():
    print("Loading and cleaning data...")
    df = pd.read_csv('crimes.csv')

    # Apply cleaning steps if 'crimes_cleaned.csv' was meant to be a cleaned version
    df['Weapon Desc'] = df['Weapon Desc'].fillna('No Weapon / Not Recorded')
    df['Vict Sex'] = df['Vict Sex'].fillna('X').replace('H', 'X')
    df['Vict Descent'] = df['Vict Descent'].fillna('X').replace('-', 'X')

    # 1. Prepare Time and Date Columns
    df['Date Rptd'] = pd.to_datetime(df['Date Rptd'])
    df['DATE OCC'] = pd.to_datetime(df['DATE OCC'])

    # Extract the hour (e.g., 1630 becomes 16)
    df['Hour'] = (df['TIME OCC'] // 100).astype(int)

    # Extract Month and Day of the Week
    df['Month'] = df['DATE OCC'].dt.month_name()
    df['Day_of_Week'] = df['DATE OCC'].dt.day_name()

    # Calculate Reporting Lag in days
    df['Reporting_Lag'] = (df['Date Rptd'] - df['DATE OCC']).dt.days

    print("Data prepped! Generating your standing reports...")

    # ==========================================
    # GRAPH 1: Peak Hours (When do most crimes happen?)
    # ==========================================
    plt.figure(figsize=(8, 12))  # Standing orientation
    hour_counts = df['Hour'].value_counts().sort_index()

    sns.barplot(y=hour_counts.index, x=hour_counts.values, orient='h', palette='magma', hue=hour_counts.index, legend=False)
    plt.title('Crime Volume by Hour of the Day\n(Military Time)', fontsize=16, weight='bold')
    plt.ylabel('Hour of the Day (0-23)', fontsize=12)
    plt.xlabel('Number of Crimes', fontsize=12)
    plt.tight_layout()
    plt.savefig('temporal_1_peak_hours_standing.png')
    plt.close()

    # ==========================================
    # GRAPH 2: Day of the Week Trends
    # ==========================================
    plt.figure(figsize=(8, 12))  # Standing orientation
    days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
    day_counts = df['Day_of_Week'].value_counts().reindex(days_order)

    sns.barplot(y=day_counts.index, x=day_counts.values, orient='h', palette='viridis', hue=day_counts.index, legend=False)
    plt.title('Crime Volume by Day of the Week', fontsize=16, weight='bold')
    plt.ylabel('Day of the Week', fontsize=12)
    plt.xlabel('Total Crimes Recorded', fontsize=12)
    plt.tight_layout()
    plt.savefig('temporal_2_weekly_trends_standing.png')
    plt.close()

    # ==========================================
    # GRAPH 3: Reporting Lags (When are police notified?)
    # ==========================================
    plt.figure(figsize=(8, 12))  # Standing orientation
    # For a readable chart, we filter to crimes reported within the first 30 days
    # (Since some crimes like identity theft can have lags of years!)
    short_lag_df = df[df['Reporting_Lag'] <= 30]

    sns.histplot(data=short_lag_df, y='Reporting_Lag', bins=31, color='crimson')
    plt.title('Reporting Lag: Days Between Crime & Police Report\n(First 30 Days)', fontsize=16, weight='bold')
    plt.ylabel('Days Delayed Before Reporting', fontsize=12)
    plt.xlabel('Number of Incidents', fontsize=12)
    plt.tight_layout()
    plt.savefig('temporal_3_reporting_lag_standing.png')
    plt.close()

    # Print out a quick statistical summary to the terminal
    print("\n--- TEMPORAL INSIGHTS SUMMARY ---")
    print(f"Most dangerous hour: {hour_counts.idxmax()}:00")
    print(f"Most dangerous day: {day_counts.idxmax()}")
    print(f"Average reporting lag: {df['Reporting_Lag'].mean():.1f} days")
    print(f"Median reporting lag: {df['Reporting_Lag'].median():.0f} days")
    print("---------------------------------")
    print("Success! Check your folder for the 3 standing PNG images.")

if __name__ == "__main__":
    main()
```

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set a clean visual style
sns.set_theme(style="whitegrid")

def main():
    print("Loading cleaned data...")
    df = pd.read_csv('crimes.csv')

    # Ensure Weapon Desc is properly tracked for "Weapon Hotspots"
    # Create a boolean column: True if a weapon was used, False if not
    df['Weapon_Used'] = df['Weapon Desc'] != 'No Weapon / Not Recorded'

    print("Data prepped! Generating the analytical charts...")

    # ==========================================
    # QUESTION 2: SPATIAL ANALYSIS (Where to deploy)
    # ==========================================

    # 2A. Area Hotspots (Total Crimes by Area)
    plt.figure(figsize=(8, 14))
    area_counts = df['AREA NAME'].value_counts()
    sns.barplot(y=area_counts.index, x=area_counts.values, orient='h', palette='Reds_r', hue=area_counts.index, legend=False)
    plt.title('Crime Hotspots:\nTotal Volume by LAPD Division', fontsize=16, weight='bold')
    plt.ylabel('LAPD Patrol Division', fontsize=12)
    plt.xlabel('Number of Crimes', fontsize=12)
    plt.tight_layout()
    plt.savefig('spatial_1_hotspots.png')
    plt.close()

    # 2B. Weapon Hotspots (Areas with highest weapon usage)
    plt.figure(figsize=(8, 12))
    weapon_areas = df[df['Weapon_Used']]['AREA NAME'].value_counts()
    sns.barplot(y=weapon_areas.index, x=weapon_areas.values, orient='h', palette='Oranges_r', hue=weapon_areas.index, legend=False)
    plt.title('Weapon Hotspots:\nIncidents Involving Weapons by Division', fontsize=16, weight='bold')
    plt.ylabel('LAPD Patrol Division', fontsize=12)
    plt.xlabel('Number of Armed Incidents', fontsize=12)
    plt.tight_layout()
    plt.savefig('spatial_2_weapon_hotspots.png')
    plt.close()

    # ==========================================
    # 3A. Top 15 Overall Crimes (LANDSCAPE)
    # ==========================================
    plt.figure(figsize=(14, 8))  # Landscape dimensions (Width: 14, Height: 8)
    top_crimes = df['Crm Cd Desc'].value_counts().head(15)

    # Swapped x and y to make the bars vertical
    sns.barplot(x=top_crimes.index, y=top_crimes.values, palette='Blues_r', hue=top_crimes.index, legend=False)

    plt.title('Most Frequent Crimes in Los Angeles\n(Top 15)', fontsize=16, weight='bold')
    plt.xlabel('Crime Description', fontsize=12)
    plt.ylabel('Number of Occurrences', fontsize=12)

    # Rotate the x-axis labels 45 degrees so the long crime names fit side-by-side
    plt.xticks(rotation=45, ha='right')

    plt.tight_layout()
    plt.savefig('victim_1_top_crimes_landscape.png')
    plt.close()
    print("- Saved victim_1_top_crimes_landscape.png")

    # ==========================================
    # 3B. Victim Age Distribution by Top 5 Crimes (LANDSCAPE)
    # ==========================================
    plt.figure(figsize=(14, 8))  # Landscape dimensions
    top_5_crime_names = df['Crm Cd Desc'].value_counts().head(5).index
    df_top_5_crimes = df[df['Crm Cd Desc'].isin(top_5_crime_names)]

    # Swapped x and y to make the boxplots vertical
    sns.boxplot(x='Crm Cd Desc', y='Vict Age', data=df_top_5_crimes, palette='Set2', hue='Crm Cd Desc', legend=False)

    plt.title('Victim Age Demographics\nAcross Top 5 Crimes', fontsize=16, weight='bold')
    plt.xlabel('Crime Category', fontsize=12)
    plt.ylabel('Victim Age', fontsize=12)

    # Rotate the x-axis labels to prevent overlapping
    plt.xticks(rotation=15, ha='right')

    plt.tight_layout()
    plt.savefig('victim_2_age_demographics_landscape.png')
    plt.close()
    print("- Saved victim_2_age_demographics_landscape.png")
    # Print summary to terminal
    print("\n--- SPATIAL & VICTIMOLOGY SUMMARY ---")
    print(f"Most dangerous division overall: {area_counts.idxmax()}")
    print(f"Division with most weapon incidents: {weapon_areas.idxmax()}")
    print(f"Most common crime in LA: {top_crimes.idxmax()}")
    # print(f"Division needing most investigative help: {unresolved.idxmax()}") # Removed undefined variable 'unresolved'
    print("---------------------------------------")
    print("Success! Check your folder for the 5 new PNG images.")

if __name__ == "__main__":
    main()
```

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set a clean visual style
sns.set_theme(style="whitegrid")

def main():
    print("Loading cleaned data...")
    df = pd.read_csv('crimes.csv')

    # Extract the hour from 'TIME OCC'
    df['Hour'] = (df['TIME OCC'] // 100).astype(int)

    print("Calculating the Deployment Matrix...")

    # Create a cross-tabulation (matrix) of Area vs. Hour
    deployment_matrix = pd.crosstab(df['AREA NAME'], df['Hour'])

    # ==========================================
    # 4. PATROL DEPLOYMENT HEATMAP
    # ==========================================
    # Set figure size to a tall, vertical orientation to cleanly list all divisions
    plt.figure(figsize=(10, 14))

    # Generate the heatmap
    # Using 'YlOrRd' (Yellow-Orange-Red) to naturally highlight the highest risk zones
    sns.heatmap(deployment_matrix, cmap='YlOrRd', linewidths=0.5,
                cbar_kws={'label': 'Volume of Crimes', 'orientation': 'horizontal', 'pad': 0.05})

    plt.title('LAPD Patrol Deployment Matrix\n(Crime Density by Division and Hour)',
              fontsize=16, weight='bold', pad=20)
    plt.ylabel('LAPD Patrol Division', fontsize=12, weight='bold')
    plt.xlabel('Hour of the Day (0-23)', fontsize=12, weight='bold')

    # Ensure the Y-axis labels are easy to read
    plt.yticks(rotation=0)

    plt.tight_layout()
    plt.savefig('4_deployment_matrix_standing.png', dpi=300)
    plt.close()

    # Identify the absolute highest-risk shift overall
    max_area = deployment_matrix.max(axis=1).idxmax()
    max_hour = deployment_matrix.loc[max_area].idxmax()

    print("\n--- DEPLOYMENT RECOMMENDATION SUMMARY ---")
    print(f"Absolute Peak Deployment Target:")
    print(f"Area: {max_area}")
    print(f"Hour: {max_hour}:00")
    print("-----------------------------------------")
    print("Success! Check your folder for 4_deployment_matrix_standing.png.")

if __name__ == "__main__":
    main()
```
